```python
import requests
from bs4 import BeautifulSoup
import html5lib
import selenium
from selenium import webdriver
import pandas as pd
```

```python
# parsing journal url
journal_url = []
url_17_36 = "https://kcginfo.kcg.gov.tw/Publish.aspx?Year=1&n=A22859B204186560&page=2&PageSize=20"
url_1_16 = "https://kcginfo.kcg.gov.tw/Publish.aspx?Year=1&n=A22859B204186560&page=3&PageSize=20"
driver = webdriver.Chrome('./chromedriver.exe')
driver.get(url_17_36)
elems = driver.find_elements_by_xpath('//a[@title="Maritime Capital Kaohsiung"]')
for elem in elems:
    journal_url.append(elem.get_attribute('href'))
driver.close()
driver = webdriver.Chrome('./chromedriver.exe')
driver.get(url_1_16)
elems = driver.find_elements_by_xpath('//a[@title="Maritime Capital Kaohsiung"]')
for elem in elems:
    journal_url.append(elem.get_attribute('href'))
driver.close()
```

```python
article_url = []
def get_article_url(url):
    driver = webdriver.Chrome('./chromedriver.exe')
    driver.get(url)
    div = driver.find_element_by_xpath('//div[@id="ctl00_ContentPlaceHolder1_YearAndChapter"]')
    elems = div.find_elements_by_tag_name('a')
    for elem in elems:
        article_url.append(elem.get_attribute('href'))
    driver.close()
```

```python
for i in journal_url:
    get_article_url(i)
```
